# 예측 유사도 확인

```python
for order in itertools.combinations(range(len(total_csv)), 2):
    merge_csv = pd.merge(total_csv[order[0]][1], total_csv[order[1]][1], on = 'id')
    merge_csv['same'] = merge_csv.iloc[:,1] == merge_csv.iloc[:,2]
    print(f"{total_csv[order[0]][0]} - {total_csv[order[1]][0]}\n   유사도: ", len(merge_csv[merge_csv['same']==True]) / len(merge_csv))
```

# csv merge 함수

```python
def merge_weights(weight, total_csv):
    for i, csv in enumerate([total_csv[i][1] for i in range(len(total_csv))]):
        for j , _ in enumerate(range(weight[i])):
            if i == 0 and j == 0:
                merge_weight_csv = csv
            else:
                merge_weight_csv = pd.merge(merge_weight_csv, csv, on = 'id')
    return merge_weight_csv
```

```python
def mode_cal(temp_merge):
    temp_merge['mode'] = temp_merge.iloc[:,1:].mode(axis = 1)[0]
    temp_merge = temp_merge[['id', 'mode']]
    temp_merge.rename(columns = {'mode':'label'}, inplace = True)
    return temp_merge
```

# ensemble 비율 조정

    - csv1 | csv2 | csv3  
    '훼손'| '훼손'| '오염'
    - input: 1, 1, 1 -> '훼손'
    - input: 1, 1, 3 -> '오염'

```python
while True:
    weight = list(map(int, input('가중치를 ,로 구분하여 입력하세요.\n').split(',')))
    if len(weight) != len(total_csv):
        weight = list(map(int, input('csv의 수와 가중치의 수가 다릅니다. 다시 입력하세요.\n').split(',')))
    else:
        break
```

```python
temp_merge = merge_weights(weight, total_csv)
temp_merge
```

```python
final_csv = mode_cal(temp_merge)
final_csv # 53 58 49 53
```
